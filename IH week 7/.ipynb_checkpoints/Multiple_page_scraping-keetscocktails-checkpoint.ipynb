# Find out how to now run scraping on all URL´s

```python
presi_urls=soup.select('td:nth-child(3) > b > a')
```

```python
presi_urls[0]['href']
```

```python
len(presi_urls)
```

```python
links=[]
for p in presi_urls:
    href1=p['href']
    url='https://en.wikipedia.org'+href1
    links.append(url)
```
