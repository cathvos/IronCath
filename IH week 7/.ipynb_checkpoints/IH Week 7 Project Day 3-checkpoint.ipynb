# clustering

```python
import pandas as pd 
import matplotlib.pyplot as plt
import seaborn as sns 
from sklearn.cluster import KMeans
```

```python
df = pd.DataFrame (df_comp, columns = ['danceability','energy','key','loudness',
                                       'mode','speechiness','acousticness','instrumentalness',
                                       'liveness','valence','tempo'])
df
```

```python
from sklearn.preprocessing import MinMaxScaler
scaler=MinMaxScaler()
scaler.fit(df)
```

```python
scaled_df=scaler.transform(df)
```

```python
df.columns
```

```python
scaled_df_music=pd.DataFrame(scaled_df,columns=['danceability', 'energy', 'key', 'loudness', 'mode', 'speechiness',
       'acousticness', 'instrumentalness', 'liveness', 'valence', 'tempo'])
```

```python
kmeans=KMeans(n_clusters=30, random_state=5)
```

```python
cluster_scaled=kmeans.fit(scaled_df_music)
```

```python
scaled_df_music
```

```python
cluster_scaled.cluster_centers_
```

```python
cluster_scaled.labels_
```

```python
scaled_df_music['cluster_scaled']=cluster_scaled.labels_
```

```python
scaled_df_music['cluster_scaled'].value_counts()
```

```python
scaled_df_music['cluster']=cluster_scaled.labels_
```

```python
scaled_df_musicdrop(['cluster_scaled'], axis = 1, inplace = True) 
```

```python
scaled_df_music
```

# pick the right number of clusters

```python
cluster_scaled.inertia_
```

```python
K=range(15,40) # whats the range of k that makes sense for your data ? 
inertia=[]
for k in K:
    kmeans=KMeans(n_clusters=k,random_state=5)
    kmeans.fit(scaled_df_music)
    inertia.append(kmeans.inertia_)
```

```python
import numpy as np 
```

```python
#plot the inertia: 
plt.figure(figsize=(16,8))
plt.plot(K,inertia,'bx-')
plt.xlabel('k')
plt.ylabel('inertia')
plt.xticks(np.arange(min(K),max(K)+1,1.0))
plt.title('elbow method - finding the optimal k by inertia');
```
