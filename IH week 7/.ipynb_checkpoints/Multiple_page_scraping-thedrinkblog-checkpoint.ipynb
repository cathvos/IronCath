```python
url_2='https://thedrinkblog.com/recipes/'
```

```python
ua = 'Mozilla/5.0 (Windows; U; Windows NT 5.1; en-US; rv:1.9.0.7) Gecko/2009021910 Firefox/3.0.7'
response = requests.get(url_2, headers={"User-Agent": ua})
response.status_code
```

```python
soup=BeautifulSoup(response.content, 'html.parser')
```

```python
print(soup.prettify())
```

```python
h4_tags=soup.find_all('h4')
```

```python
names=[]
for i in h4_tags:
    names.append(i.get_text())
```

```python
names
```

```python
converted_nameslist = []

for element in names:
    converted_nameslist.append(element.strip())

print(converted_nameslist)
```

```python
nameslist = []

for element in converted_nameslist:
    nameslist.append(element.replace(" ", "-"))

print(nameslist)
```

```python
nameslist2 = []

for element in nameslist:
    nameslist2.append(element.replace("’", ""))

print(nameslist2)
```

```python
nameslist3 = []

for element in nameslist2:
    nameslist3.append(element.replace('Corpse-Reviver-#2', 'Corpse-Reviver-2'))

print(nameslist3)
```

```python
nameslist_final = []

for element in nameslist3:
    nameslist_final.append(element.replace("Peño-Blast", "the-peno-blast"))

print(nameslist_final)
```

```python
for i in range(len(nameslist_final)):
    nameslist_final[i] = nameslist_final[i].lower()

print(nameslist_final)
```

```python
len(converted_nameslist)
```

```python
links=[]
for i in nameslist_final:
    href1=i
    url='https://thedrinkblog.com/'+href1+'/'
    links.append(url)
```
