### Create Python function named rentals_may to connect to the Sakila database 
### and return dataframe storing the data from rental table with information for May 2005.

```python
def rentals_may():
    engine=create_engine(connection_string)
    df=pd.read_sql_query("select * from rental WHERE rental_date > '2005-05-01%%' AND rental_date < '2005-05-31%%';",engine)
    return df
```

```python
rentals_may()
```

```python
df_rentals_may= rentals_may()
```

```python
df_rentals_may
```

### Create Python function named rentals_june to connect to the Sakila database 
### and return dataframe storing the data from rental table with information for June 2005.

```python
def rentals_june():
    engine=create_engine(connection_string)
    df=pd.read_sql_query("select * from rental WHERE rental_date >= '2005-06-01%%' AND rental_date <= '2005-06-30%%';",engine)
    return df
```

```python
rentals_june()
```

```python
df_rentals_june= rentals_june()
```

```python
df_rentals_june
```

### Create a Python function to get the number of rentals for each customer in May given 
### the dataframe provided by the function rentals_may. 
### Hint: Consider making use of pandas groupby() and pandas agg()

```python
def rentals_count_may2():
    data_may2=df_rentals_may.groupby("customer_id", as_index=False).agg({"rental_id":"count"})
    return data_may2
```

```python
rentals_count_may2()
```

```python
df_rentals_may_count= rentals_count_may2() # creating df from function
```

```python
df_rentals_may_count.rename(columns={'rental_id': 'count_may'},inplace=True, errors='raise') #renaming the column
```

```python
df_rentals_may_count
```

### Create a Python function to get the number of rentals for each customer in June 
### given the dataframe provided by the function rentals_june.

```python
def rentals_count_june():
    data_june=df_rentals_june.groupby("customer_id", as_index=False).agg({"rental_id":"count"})
    return data_june
```

```python
rentals_count_june()
```

```python
df_rentals_june_count= rentals_count_june() # creating df from function
```

```python
df_rentals_june_count.rename(columns={'rental_id': 'count_june'},inplace=True, errors='raise') #renaming the column
```

```python
df_rentals_june_count
```

### Write a Python function named check_rentals that given:
### - a dataframe with the number of rentals in May by each customer_id
### - a dataframe with the number of rentals in June by each customer_id return a new dataframe with the following columns:
#### - customer_id
#### - Rentals_May
#### - Rental_June
#### - The difference between the previous columns
### Hint: For this part, you can create a join between the two dataframes created before, using the merge function 
### available for pandas dataframes. Here is a link to the documentation for the merge function.

```python
def check_rentals():
    df_cr=pd.merge(df_rentals_may_count, df_rentals_june_count, how='left', left_on = 'customer_id', right_on = 'customer_id')
    df_cr['diff_may_june'] = df_cr['count_june'] - df_cr['count_may']
    return df_cr
```

```python
check_rentals()
```

```python
df_diff=check_rentals()
```

```python
df_diff
```

### Create a Python function that given the dataframe returned by the previous function, returns another 
### dataframe with the customer_id of customers who rented movies in may but didn't rent any movie in June.

```python
def rentals_mayyes_juneno():
    df_diff_nulls = df_diff.query('count_june.isnull()') # add .drop columns count and diff
    return df_diff_nulls
```

```python
rentals_mayyes_juneno()
```

```python
df_diff_new=rentals_mayyes_juneno()
```

```python
df_diff_new= df_diff_new.drop(columns=['count_may', 'count_june', 'diff_may_june'])
```
