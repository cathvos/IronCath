```python
# code to change different discriptions in a column to a limited number of options

gender_list = [] #create an empty list where youÂ´ll append the new items to.
for i in range(len(health_2["gender"])):
    if health_2["gender"][i] == "male" or health_2["gender"][i] == "Male" or health_2["gender"][i] == "M":
        gender_list.append("male")
    elif health_2["gender"][i] == "female" or health_2["gender"][i] == "Female" or health_2["gender"][i] == "F"or health_2["gender"][i] == "feamale":
        gender_list.append("female")
    else:
        gender_list.append("unknown")
```
