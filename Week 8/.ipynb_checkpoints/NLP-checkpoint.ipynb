# 1. Acitivity 1 - NLTK on news article

```python
import docx2txt
import nltk
```

```python
nltk.download()
```

# 1.1 get chosen text

```python
mytext=docx2txt.process('Horseracing.docx')
```

```python
mytext
```

# 1.2 nltk tools to process the text

```python
from nltk.tokenize import word_tokenize
```

```python
mytext=word_tokenize(mytext)
```

```python
mytext
```

```python
len(mytext)
```

```python
# lets put everything in lower case
lower_text=[word.lower() for word in mytext]
```

```python
# lets count  the frequency of words
fdist_v1=nltk.FreqDist(lower_text)
```

```python
word_freq_v1=dict((word, freq)for word, freq in fdist_v1.items())
```

```python
# sort it by the value freq
sorted(word_freq_v1.items(), key=lambda x:x[1], reverse=True)
```

# 1.3 text cleaning
### - remove stop words
### - remove punctuation
### - lemmatising / stemming
### - pronouns, nouns and entities/ abbreviations
### - prepositions

```python
# punctuation spaces numbers - isalpha
clean_text=[word for word in lower_text if word.isalpha()]
```

```python
len(clean_text)
```

```python
# get stop words - appropriate for text
from nltk.corpus import stopwords
```

```python
my_stopwords=stopwords.words('english')
```

```python
my_stopwords
```

```python
clean_text=[word for word in clean_text if word not in my_stopwords]
```

```python
len(clean_text)
```

# version 2

```python
# lets count  the frequency of words
fdist_v2=nltk.FreqDist(clean_text)
```

```python
word_freq_v2=dict((word, freq)for word, freq in fdist_v2.items())
```

```python
# sort it by the value freq
sorted(word_freq_v2.items(), key=lambda x:x[1], reverse=True)
```

```python
porter=nltk.PorterStemmer()
```

```python
stem_words=[porter.stem(word) for word in clean_text]
```

```python
len(stem_words)
```

```python
# 3. third version
```

```python
# lets count  the frequency of words
fdist_v3=nltk.FreqDist(stem_words)
```

```python
word_freq_v3=dict((word, freq)for word, freq in fdist_v3.items())
```

```python
# sort it by the value freq
sorted(word_freq_v3.items(), key=lambda x:x[1], reverse=True)
```

# 5. Activity 2

```python
import pandas as pd
from textblob import TextBlob
```

```python
clothingdf=pd.read_csv('Womens Clothing E-Commerce Reviews.csv')
```

```python
clothingdf.info()
```

```python
clothingdf=clothingdf[['Age', 'Title', 'Review Text', 'Rating', 'Division Name', 'Department Name']]
```

```python
clothingdf.info()
```

```python
# sentiment
TextBlob("i hate this. it is disgusting").sentiment
# sentiment polarity -1 (neg) to +1 (pos)
# sentiment subjectivity 0 (obj) to 1 (subj)
```

```python
clothingdf['Rating'].value_counts()
```

```python
clothingdf['Department Name'].value_counts()
```

```python
rating_by_depdiv=clothingdf.groupby(['Division Name','Department Name']).agg({'Rating':['mean', 'max', 'min', 'median']})
```

```python
rating_by_depdiv
```

# 5.1 get sentiment scores

```python
# textblob - cant handle nulls
clothingdf=clothingdf.dropna(subset=['Review Text'], how='any')
```

```python
clothingdf.iloc[34]['Review Text']
```

```python
example=TextBlob(clothingdf.iloc[34]['Review Text'])
example.sentiment
```

```python
def get_sentiment_p(text):
    return TextBlob(text).sentiment.polarity
```

```python
clothingdf['sentiment_p']=clothingdf['Review Text'].apply(get_sentiment_p)
```

```python
def get_sentiment_s(text):
    return TextBlob(text).sentiment.subjectivity
```

```python
clothingdf['subjectivity']=clothingdf['Review Text'].apply(get_sentiment_s)
```

```python
clothingdf.head()
```

# 5.2 analysis - avg rating, sentiment, subjectivity per dept div

```python
ratingsensub_by_dep=clothingdf.groupby(['Division Name','Department Name']).agg({'Rating':['mean'], 'sentiment_p':['mean', 'min', 'max'], 'subjectivity':['mean']})
```

```python
ratingsensub_by_dep
```

# 6. Spacy Example 3

```python
import spacy
nlp =spacy.load("en_core_web_sm")
```

```python
tweets=pd.read_csv('ever_trump.csv')
```

```python
tweets.head()
```

```python
tokens=nlp(''.join(str(tweets.text.tolist())))
```

```python
len(tokens)
```

```python
from collections import Counter
```

```python
items=[x.text for x in tokens.ents] # finds all entities, ignores stopwords and punctuation
Counter(items).most_common(20)
```

```python
person_list=[]
for ent in tokens.ents:
    if ent.label_=='PERSON':
        person_list.append(ent.text)
```

```python
person_list
```

```python
person_counts=Counter(person_list).most_common(20)
df_person=pd.DataFrame(person_counts, columns=['text', 'count'])
df_person.plot.barh(x='text', y='count');
```
